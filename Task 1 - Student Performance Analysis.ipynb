# Visualiation

```python
plt.hist(student_data['G3'], bins = 10, color = 'seagreen', edgecolor='darkblue')
plt.title("Overall Insights of final grades")
plt.xlabel("Grades Review")
plt.ylabel("List of Students")
plt.show()
```

## Distribution of Final Grades
- The histogram displays that most students are scored between 8 and 14 marks
- The highest concentration of students is around 10 to 12 marks
- Very few students acchieved a very high or very low
- This represents that student performance is mostly average and follows a near normal distribution

```python
#Study Time vs Grades

plt.scatter(student_data['studytime'], student_data['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
```

## Study Time vs Final Grade
- The scatter plot displays a broader circulation of data points
- There is vague upward or downward trend
- Students with equivalent study time own diverse grades
- This represents a poor relationship between study time and performance
- Study time alone does not strongly influence final grades

```python
# male vs female average score

student_data.groupby('sex')['G3'].mean().plot(kind='bar')

plt.title("Standard Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Final Grade")
plt.show()
```

## Gender-based Performance Visualization
- The bar chart displays the standard conclusion grade for both male and female students
- Male students have a slightly higher average score compared to female students
- However, the difference is small, indicating similar performance levels
